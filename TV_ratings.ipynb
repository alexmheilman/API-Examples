```python
# Dependencies
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tv_shows = ["It's Always Sunny in Philadelphia", 'Workaholics', 'Abbott Elementary', 'Bleach', 'Naruto', 'Vikings', 'Stranger Things','Chainsaw Man', 'The Wire']
name = []
rating = []


url = 'https://api.tvmaze.com/search/shows?q='

for title in tv_shows:
    query_url = url + title
    show_info = requests.get(query_url).json()
    name.append(show_info[0]['show']['name'])
    rating.append(show_info[0]['show']['rating']['average'])
    
print(name)
rating
```

```python
# create DataFrame
shows_df = pd.DataFrame({'Title': name, 'Rating': rating})
shows_df
```

```python
# Plot the data
plt.bar(shows_df['Title'], shows_df['Rating'])
plt.xticks(rotation = 90)
plt.xlabel('Title')
plt.ylabel('Averge Rating')
plt.title('TV Show Ratings')
plt.show()
```
